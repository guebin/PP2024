(풀이) – 반복문 사용, if문 사용

```python
f = lambda x: 1/9*x*(x-6)*(x-9)
def g(x,t):
    if x<t:
        return f(x)
    else:
        return -(x-t)+f(t)
```

```python
x = np.linspace(0,10,101)
plt.plot(x,f(x))
plt.plot(x,[g(xi,3) for xi in x])
```

```python
t = np.linspace(0,6,101)
t
```

```python
s = lambda t: np.max(np.cumsum([g(xi,t) for xi in x])*0.1)
```

```python
plt.plot(t,[s(t) for t in t])
```

```python
np.max([s(t) for t in t])
```

```python
hx = np.cumsum(fx - gx)*0.06
plt.plot(x,hx,'--')
```
