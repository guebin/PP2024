# 문제 1. – 10점

$f(x)=x^3$ 위의 임의의 점을 입력하면 접선을 리턴하는 함수를 구현하라.
이러한 함수를 이용하여 아래와 같은 그래프를 그려라. (범례는 표현하지
않아도 감점없음)

```python
#
```

-   $x$의 범위는 $(-5,5)$ 이다.
-   두 그래프는 $x=3$ 에서 접한다.

(풀이)

```python
def make_func(x0):
    f = lambda x: x**3
    def func(x):
        h = 0.0001
        a = (f(x0+h)-f(x0))/h 
        return a*(x-x0) + f(x0)
    return func
x = np.linspace(-5,5,101)
tan_line = make_func(3)
plt.plot(x,x**3,'--')
plt.plot(x,tan_line(x),'--')
```

# 문제 7. – 30점

첫째항이 1이고 공차가 $0<d<1$ 인 등차수열에서,
$\sum_{n=1}^{75}a_n=1046.25$ 를 만족하는 $d$를 찾아라.

**힌트1:** 첫째항이 1이고 공차가 $d>0$인 등차수열의 일반항은
$a_n = 1 +(n-1)d$와 같이 표현할 수 있다.

**힌트2:** $d$는 아래중 하나의 값을 가진다.

```python
np.linspace(0,1,101)[1:-1]
```

(풀이)

```python
def func(d):
    n = np.arange(1,76)
    an = 1+(n-1)*d
    return np.abs(np.sum(an)-1046.25)
```

```python
d_arr = np.linspace(0,1,101)[1:-1]
d_arr[np.argmin(list(map(func,d_arr)))]
```
